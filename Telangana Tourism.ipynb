```python
import pandas as pd
```

```python
# List of CSV files to read
csv1=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/domestic_visitors/domestic_visitors_2016.csv')
csv2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/domestic_visitors/domestic_visitors_2017.csv')
csv3=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/domestic_visitors/domestic_visitors_2018.csv')
csv4=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/domestic_visitors/domestic_visitors_2019.csv')
```

```python
# Concatenate the dataframes vertically
domestic_visitors = pd.concat([csv1,csv2,csv3,csv4], ignore_index=True)

```

```python
domestic_visitors.tail(50)
```

```python
domestic_visitors.isnull().sum()
```

```python
# Identify rows with null values
null_rows = domestic_visitors[domestic_visitors.isnull().any(axis=1)]

# Display the null rows
print(null_rows)
```

```python
dm_visitors=domestic_visitors.copy()
```

```python
dm_visitors.isnull().sum()
```

```python
dm_visitors=dm_visitors.dropna() # to remove null values
```
